```python
import pandas as pd
import tensorflow as tf

import autokeras as ak
```

```python
# traindf = pd.read_csv('/kaggle/input/clean-creditcard-train-data/clean-credit-train.csv')
# testdf = pd.read_csv('/kaggle/input/clean-creditcard-test-data/clean-credit-test.csv')
# traindf = tf.keras.utils.get_file("clean-credit-train.csv", "https://www.kaggle.com/datasets/mihikajain/clean-creditcard-train-data")

traindf = "/kaggle/input/clean-creditcard-train-data/clean-credit-train.csv"
```

```python
clf = ak.StructuredDataClassifier(
    overwrite=True, max_trials=8
)
clf.fit(
    # The path to the train.csv file.
    traindf,
    # The name of the label column.
    "Class",
    epochs=8,
)
```

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import load_model
```
